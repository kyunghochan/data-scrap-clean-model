```python
df_1 = pd.read_csv('/content/drive/MyDrive/SKRIPSI/dataset/labeled/df_1.csv')
pd.set_option("max_colwidth", None)
df_1.tail()
```

```python
df_1 = df_1.drop(['author_name'], axis=1)
df_1.tail()
```

```python
df_1['original_text'] = df_1['text']
```

```python
df_1['sentiment'].unique()
```

```python
df_1['sentiment'].value_counts()
```

## remove emoji

```python
def remove_emojis(text):
    emoji = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642"
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoji, '', text)
```

```python
df_1['text'] = df_1['text'].apply(remove_emojis)
df_1.tail()
```

## remove html tag

```python
df_1['text'] = df_1['text'].str.replace(r'<[^<>]*>', '', regex=True)
df_1.tail()
```

## fix typo dict (filtering)

```python
with open('/content/drive/MyDrive/SKRIPSI/corrections json/df_1_json.json', 'r') as f:
    corrections = json.load(f)

def correct_text(text):
    words = text.split()
    corrected_words = []

    for word in words:
        if word in corrections:
            corrected_words.append(corrections[word])
        else:
            corrected_words.append(word)

    corrected_text = ' '.join(corrected_words)
    return corrected_text

df_1['text'] = df_1['text'].apply(correct_text)
df_1.tail()
```

## replace all (-) with space

```python
def replace_hyphens_with_spaces(text):
    return text.replace("-", " ")

df_1['text'] = df_1['text'].apply(replace_hyphens_with_spaces)
```

```python
df_1.tail()
```

## remove all punctuations/special characters & numbers

```python
def remove_punctuation_numbers(text):
    punctuation = str.maketrans('', '', string.punctuation + string.digits)
    return text.translate(punctuation)

df_1['text'] = df_1['text'].apply(remove_punctuation_numbers)
```

```python
df_1.tail()
```

## turn string into lower case (Case Folding)

```python
def string_lower(text):
  return text.lower()

df_1['text'] = df_1['text'].apply(string_lower)
```

```python
df_1.tail()
```

## remove stopword

```python
def remove_stopwords(text):
    words = word_tokenize(text)
    stopwords_id = set(stopwords.words('indonesian'))
    filtered_words = [word for word in words if word.lower() not in stopwords_id]
    filtered_text = ' '.join(filtered_words)
    return filtered_text

df_1['text'] = df_1['text'].apply(remove_stopwords)
```

```python
df_1.tail(10)
```

## stemming word (sastrawi)

```python
def stemming(text):
    stemmer = StemmerFactory().create_stemmer()
    stemmed = stemmer.stem(text)
    return stemmed

df_1['text'] = df_1['text'].apply(stemming)
```

```python
df_1.tail(10)
```

```python
df_1[(df_1['text']) == ""]
```

```python
df_1 = df_1[df_1['text'] != ""]
df_1.reset_index(drop=True, inplace=True)
```

```python
df_1.info()
```

## save dataframe

```python
# df_1.to_csv('/content/drive/MyDrive/SKRIPSI/dataset/cleaned/df_1_cleaned.csv', index=False)
```
